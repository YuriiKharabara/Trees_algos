```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
# from sys import argv

data = pd.read_csv('results.csv')
data_kruskal = data.loc[data['Method'] == 'kruskal']
data_prim_heap = data.loc[data['Method'] == 'prim_heaps']
data_prim = data.loc[data['Method'] == 'prim']

x = data_kruskal['Number of verts']
y = data_kruskal['Probability']
z = data_kruskal['Time taken']
x2 = data_prim_heap['Number of verts']
y2 = data_prim_heap['Probability']
z2 = data_prim_heap['Time taken']
x3 = data_prim['Number of verts']
y3 = data_prim['Probability']
z3 = data_prim['Time taken']


fig = plt.figure(figsize=plt.figaspect(0.25))

ax = fig.add_subplot(1, 3, 1, projection='3d')
surf = ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.1, vmin=0, vmax=150)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.set_xlabel('verts')
ax.set_ylabel('probability')
ax.set_zlabel('Time')
ax.view_init(30, 245)

ax2 = fig.add_subplot(1, 3, 2, projection='3d')
surf2 = ax2.plot_trisurf(x2, y2, z2, cmap=cm.jet, linewidth=0.1, vmin=0, vmax=150)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax2.set_xlabel('verts')
ax2.set_ylabel('probability')
ax2.set_zlabel('Time')
ax2.view_init(30, 245)

ax3 = fig.add_subplot(1, 3, 3, projection='3d')
surf3 = ax3.plot_trisurf(x3, y3, z3, cmap=cm.jet, linewidth=0.1, vmin=0, vmax=150)
fig.colorbar(surf, shrink=0.5, aspect=5)

ax3.set_xlabel('verts')
ax3.set_ylabel('probability')
ax3.set_zlabel('Time')
ax3.view_init(30, 245)


# plt.savefig('prim_diagram.pdf')
plt.show()

```

```python
data = pd.read_csv('results.csv')
for num in [10, 20, 50, 100, 200, 300, 500]:
    data_kruskal = data.loc[data['Method'] == 'kruskal'].loc[data['Number of verts'] == num]
    data_prim = data.loc[data['Method'] == 'prim_heaps'].loc[data['Number of verts'] == num]

    x = data_kruskal['Probability']
    y = data_kruskal['Time taken']
    x2 = data_prim['Probability']
    y2 = data_prim['Time taken']


    fig, ax1 = plt.subplots(1)
    fig.suptitle(f'{num} verts minimal span tree')
    ax1.plot(x, y)
    ax1.plot(x2, y2)
    ax1.grid(True)
    ax1.set_ylabel('Time')
    ax1.set_xlabel('Probability')

    plt.savefig(f'{num}_verts_diagram.pdf')
```
